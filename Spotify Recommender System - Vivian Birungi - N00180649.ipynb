## Recommender Systems

Recommender systems are a computer-based, machine learning algorithm that collect user information and return products or services the user may be interested in for example, every time a user shops online, the recommender system is taking in what they may be interested in and returning similar items. (Shetty, 2019-2021)

Recommender systems make the task of scrolling through the internet, searching for what you want, much easier and cut down on time spent looking for items as well as introducing new items to users that they may not have thought of. This increases the user’s time spent online as they are exposed to more of what they like and therefore, recommender systems collect more data from the knowledge they gain about the users.

A person’s brain releases dopamine or ‘feel good’ chemical responses when faced with things they enjoy, and this comes easy while scrolling through the internet (BetterHealth, 2021). Many applications and social media platforms use recommender systems because they play a big role in entertaining these chemical responses that keep users engaged and give a better user experience (BetterHealth, 2021).

There are many different types of recommender systems. In this chapter, both collaborative based and content-based recommender systems will be discussed.

### Collaborative Based

There are two types of collaborative based filtering systems. Collaborative based recommender systems look at the similarities in both users and items (Google Developers, 2021). 

Part of this means that users with similar interests can influence each other’s recommendations for example, if both users like Taylor Swift and one user starts listening to Macklemore, the other user may also be recommended Macklemore. Furthermore, recommendations can also be made based on what items or content the user continues to interact with most. 

The user will then be recommended similar items or content relating to what they currently like (Mohd Abdul Hameed, 2012).

#### Advantages

 - This style of recommender is good because it focuses mainly on the content/items the user has shown interests in (Dangeti, 2021).
 
 - Even if there are no items on the user has interacted with, the recommender system can still make recommendations based on what is popular to other users (Dangeti, 2021)

#### Disadvantages

 - Depends on having enough users to start engaging with items/content. They need users to interact with items to make recommendations.
 
 - Would not make recommendations based on user’s personal interests e.g., if a new user who likes horror movies keeps getting recommended comedy movies because that’s what is popular among other users on the site.

### Content Based

Content based recommender systems look at what the users’ interests are. 
They mainly focus on combining the users interests and the products/services available and providing unique recommendations suited to the user (Iterators, 2021).

The recommender also keeps track of what the user has shown interest in the most in the past and makes recommendations based on what items/services the user liked (Iterators, 2021).

#### Advantages

 - Content based recommender systems do not always require a user to directly interact with the items/content as it will mostly be looking at the user’s profile (Pasquale Lops, 2011).
 
 - They can recommend new content that has not yet received any rating from users, once again it is recommending based on what interests the user has provided (Pasquale Lops, 2011).
 
#### Disadvantages

 - Relies on a user sharing their interests within their user profile to start making recommendations. Harder for recommendations to be made to inactive users or users who have not finished setting up their profile.
 
 - Depends on new users interacting with content consistently to build up more accurate recommendations (Pasquale Lops, 2011).

```python
#nearest neighbour algorithm
#nearest neighbour algortihm is an unsupervised classifier also known as memory based system,
#because it memorizes instances and then recommends an item (a single instance), based on how quantitively similar it is to a new, incoming instance
#producer has a specification for a particular artist they are looking for, 
#the dataset has instances(items) that may match
#need to create a single test point with producers specifications, this runs and finds the similar type within the dataset
#using spotify dataset. looking to recommend artist based on features and recommending artists with features similar.
import numpy as np
import pandas as pd

import sklearn
from sklearn.neighbors import NearestNeighbors
```

```python
#importing csv file of dataset
artists = pd.read_csv('spotify_artists.csv')
```

```python
#printing out dataset
artists
```

```python
#producers specifications i.e test point
#specifications focus on artists popularity and followers.
t = [44, 42023]
```

```python
#defined a subset from the dataset
#special indexer to select only variables we need
#seeing how nearest neighbour algorithm can be used to recommend artist for this listener based on requirements
#loc accesses a group of rows and columns by label
X = artists.loc[:,('artist_popularity', 'followers')].values
```

```python
X[0:5]
#subset below that we created from the dataset
```

```python
#building nearest neighbor model by instansiating nearest neighbor object
#passing in n_neighbors=1 to tell the algorithm to search the dataset and find a single point that is nearest to the test point that we need
#fit(X) is fitting the model to the data
#nbrs is object name
nbrs = NearestNeighbors(n_neighbors=1).fit(X)
```

```python
#to find nearest neighbor we call the kneighbor function on the test point and print it out
#assign it ot the object nbrs
#nbrs.kneighbors accesses the model from the nearest neighbour function. passing in the test point t as our object.
print (nbrs.kneighbors([t]))

#first array represents the length from test point t to p.
#dtype is the type of matrix returned by the fit method used.
#second array contains value to the nearest index, this represents X.
#to find this datapoint we scroll to the number specified in the second arrayand that is the artist we recommend to the producer.
```
